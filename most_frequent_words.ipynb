```python
import pandas as pd
import wordcloud
from matplotlib import pyplot as plt
from nltk.tokenize import sent_tokenize
from nltk.tokenize import RegexpTokenizer
from nltk.corpus import stopwords

%matplotlib inline
```

```python
df = pd.read_csv("data/debate.csv", encoding = 'iso-8859-1')
clinton_df = df[df.Speaker=="Clinton"].copy()
trump_df = df[df.Speaker=="Trump"].copy()
```

```python
trump_words = '. '.join(trump_df['Text'])
trump_cloud = wordcloud.WordCloud(background_color="white").generate(trump_words)
plt.imshow(trump_cloud)
plt.axis('off')
plt.savefig('./donald_debate_words.png')
plt.show()
```

```python
clinton_words = '. '.join(clinton_df['Text'])
clinton_cloud = wordcloud.WordCloud(background_color="white").generate(clinton_words)
plt.imshow(clinton_cloud)
plt.axis('off')
plt.savefig('./hillary_debate_words.png')
plt.show()
```

```python
tokenizer = RegexpTokenizer(r'\w+')
stop = list(stopwords.words('english'))
print "clinton top words"
print pd.Series([_ for _ in tokenizer.tokenize(clinton_words.lower()) if _ not in stop]).value_counts()[:20]
print "\ntrump top words"
print pd.Series([_ for _ in tokenizer.tokenize(trump_words.lower()) if _ not in stop]).value_counts()[:20]
```
