```python
import pandas as pd
import wordcloud
from matplotlib import pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv("data/debate.csv", index_col="Line")
clinton_df = df[df.Speaker=="Clinton"].copy()
trump_df = df[df.Speaker=="Trump"].copy()

print len(clinton_df), len(trump_df)
```

```python
trump_words = ' '.join(trump_df['Text'])
trump_cloud = wordcloud.WordCloud(background_color="white").generate(trump_words)
plt.imshow(trump_cloud)
plt.axis('off')
plt.savefig('./donald_debate_words.png')
plt.show()
```

```python
clinton_words = ' '.join(clinton_df['Text'])
clinton_cloud = wordcloud.WordCloud(background_color="white").generate(clinton_words)
plt.imshow(clinton_cloud)
plt.axis('off')
plt.savefig('./hillary_debate_words.png')
plt.show()
```
